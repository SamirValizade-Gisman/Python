```python
import re
from docx import Document
import openpyxl

# Word sənədinin yolunu qeyd edin
WORD_FILE = r"C:\Users\AGS_ASC\Desktop\coordinat/Bag monitorinq 2025.docx"
EXCEL_FILE = r"C:\Users\AGS_ASC\Desktop\coordinat/coordinates.xlsx"

# "39°36'40.71"N 47°01'50.92"E" tipli koordinat cütlərini ayırmaq üçün regex
# Bu nümunə enlik (N) və uzunluq (E) üçün tipik formata uyğundur:
#  - dərəcə: \d+°
#  - dəqiqə: \d+'
#  - saniyə: \d+(\.\d+)?"
#  - istiqamət: N|S|E|W (lakin burada adətən N-E işlənir)
coordinate_pattern = re.compile(
    r"(\d+°\d+'\d+(?:\.\d+)?\"[NS])\s+(\d+°\d+'\d+(?:\.\d+)?\"[EW])"
)

# Word sənədini oxuyuruq
doc = Document(WORD_FILE)

# Excel faylı yarat (openpyxl ilə)
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Coordinates"

# Sütun başlıqları yazaq (istəsəniz dəyişə bilərsiniz)
ws.cell(row=1, column=1, value="Latitude (Enlik)")
ws.cell(row=1, column=2, value="Longitude (Uzunluq)")

excel_row = 2  # Koordinatlar üçün yazacağımız növbəti sətir

# Word sənədində bütün cədvəlləri gəzək
for table in doc.tables:
    for row in table.rows:
        # Hər sətirdəki xanaların mətnini toplayırıq
        cell_texts = [cell.text for cell in row.cells]
        
        # Məs: "Ərazinin koordinatları 39°36'40.71"N 47°01'50.92"E" sətri
        # adətən bir xanada ola bilər, bəzən isə iki xanaya bölünə bilər və s.
        # Buna görə bütün xanalardakı mətnləri birləşdiririk:
        combined_text = " ".join(cell_texts)
        
        # "Ərazinin koordinatları" sözləri keçirsə, koordinatları axtarırıq
        if "Ərazinin koordinatları" in combined_text:
            # Regex ilə bütün koordinat cütlüklərini tapırıq
            matches = coordinate_pattern.findall(combined_text)
            
            # Hər enlik-uzunluq cütlüyü üçün Exceldə bir sətir yaz
            for (lat, lon) in matches:
                ws.cell(row=excel_row, column=1, value=lat)
                ws.cell(row=excel_row, column=2, value=lon)
                excel_row += 1

# Tapılan koordinatları Excel-ə yazmağı bitirdikdən sonra faylı saxlayırıq
wb.save(EXCEL_FILE)
print("Koordinatlar uğurla '{}' faylına yazıldı.".format(EXCEL_FILE))
```

```python
import re
from docx import Document
import openpyxl

# Fayl yolları (lazım gəldikcə dəyişdirin)
WORD_FILE = r"C:\Users\AGS_ASC\Desktop\coordinat\Bag monitorinq 2025.docx"
OUTPUT_WORD_FILE = r"C:\Users\AGS_ASC\Desktop\coordinat\Bag monitorinq 2025_UPDATED.docx"
EXCEL_FILE = r"C:\Users\AGS_ASC\Desktop\coordinat1\coordinates1.xlsx"

# Koordinatları tapmaq üçün regex (ənlik (N) + uzunluq (E) formatı)
coordinate_pattern = re.compile(
    r"(\d+°\d+'\d+(?:\.\d+)?\"[NS])\s+(\d+°\d+'\d+(?:\.\d+)?\"[EW])"
)

# Word faylını oxu
doc = Document(WORD_FILE)

# Excel faylı yarad (openpyxl)
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Coordinates"

# Excel-in başlıq sətirində sütun adları
ws.cell(row=1, column=1, value="ID")
ws.cell(row=1, column=2, value="Latitude (Enlik)")
ws.cell(row=1, column=3, value="Longitude (Uzunluq)")

excel_row = 2          # Exceldə yazmağa başlayacağımız sətir
orchard_index = 0      # “Növü” aşkar etdiyimiz hər cədvələ veriləcək ID

# Word-dəki bütün cədvəlləri iterasiya edirik
for table in doc.tables:
    # Bu cədvəldə “Növü” sözü olub-olmadığını yoxlamaq üçün
    combined_table_text = " ".join(cell.text for row in table.rows for cell in row.cells)

    # Əgər bu cədvəldə “Növü” sözü varsa, deməli, buranı nömrələyirik
    if "Növü" in combined_table_text:
        orchard_index += 1  # Yeni cədvəl üçün ID-ni artır

        # 1) Word sənədinin özündə “Növü” yazılan xanada ID-ni yerləşdirmək
        for row in table.rows:
            # Hər sətrin xanasına baxırıq, adətən [0] = "Növü", [1] = "Üzüm" və s. olur
            cell_texts = [cell.text.strip() for cell in row.cells]
            if len(cell_texts) >= 2:
                # Məsələn, birinci xananın mətni "Növü" olarsa,
                # ikinci xananı ID ilə əvəzləyirik
                if cell_texts[0] == "Növü":
                    # Məsələn, ikinci xanaya sadəcə ID yazırıq
                    row.cells[1].text = str(orchard_index)
                    break

        # 2) Eyni cədvəldəki bütün koordinat cütlərini (N/E) tap
        matches = coordinate_pattern.findall(combined_table_text)

        # 3) Tapılan koordinat cütlərini Excel-ə yaz
        for (lat, lon) in matches:
            ws.cell(row=excel_row, column=1, value=orchard_index)
            ws.cell(row=excel_row, column=2, value=lat)
            ws.cell(row=excel_row, column=3, value=lon)
            excel_row += 1

# Excel faylını saxlayırıq
wb.save(EXCEL_FILE)

# Yenilənmiş Word faylını ayrıca saxlayırıq (Növü sözünün qarşısına ID yazıldı)
doc.save(OUTPUT_WORD_FILE)

print("Word faylı yenilənərək '{}' olaraq saxlanıldı.".format(OUTPUT_WORD_FILE))
print("Koordinatlar uğurla '{}' faylına yazıldı.".format(EXCEL_FILE))
```
